## PDF_PMF_ CDF_ PPF ##

В теории вероятностей и статистике существуют четыре ключевые функции, которые описывают распределения случайных величин.

### 1. **PDF (Probability Density Function) - Функция плотности вероятности**

PDF описывает непрерывные распределения - она показывает плотность вероятности в каждой точке, а не саму вероятность.

PDF(x) = плотность вероятности в точке x

 - Значения могут быть > 1
 - Интеграл от PDF = 1
 - Вероятность = площадь под кривой PDF
 - Используется для непрерывных распределений
 
### 2. **PMF (Probability Mass Function) - Функция массы вероятности**

PMF описывает дискретные распределения - она показывает прямую вероятность того, что дискретная случайная величина примет конкретное значение.

PMF(k) = P(X = k)

- Значения всегда от 0 до 1
- Сумма всех PMF = 1
- Используется для дискретных распределений

### 3. **CDF (Cumulative Distribution Function) - Функция распределения**

CDF показывает накопленную вероятность - вероятность того, что случайная величина будет меньше или равна заданному значению.


CDF(x) = P(X ≤ x)

Связь с другими функциями:

Для дискретных: CDF(x) = Σ[k≤x] PMF(k)
Для непрерывных: CDF(x) = ∫[-∞,x] PDF(t) dt

 - Монотонно возрастает от 0 до 1
 - CDF(∞) = 1, CDF(-∞) = 0
 - Работает и для дискретных, и для непрерывных
 
### 4. **PPF (Percent Point Function) - Процентная функция**

PPF - обратная функция к CDF - она показывает значение случайной величины, соответствующее заданной вероятности (квантиль).

Если CDF(x) = p, то PPF(p) = x

- Расчет доверительных интервалов

- Генерация случайных чисел

- Определение критических значений


### Практическое применение

1. PDF/PMF:

 - Моделирование распределений данных
 - Оценка вероятностей попадания в интервалы
 - Визуализация формы распределения

2. CDF:

 - Расчет вероятностей P(X ≤ x)
 - Построение графиков вероятностей
 - Статистические тесты

3. PPF:

 - Генерация случайных выборок
 - Расчет доверительных интервалов
 - Определение критических значений для тестов
 
Например, применение в медицинской статистике:

- PMF: Вероятность 3 случаев заболевания в день
- PDF: Плотность вероятности роста пациента
- CDF: Вероятность роста ≤ 175 см
- PPF: Рост, ниже которого 90% пациентов

**PDF показывает, как изменяется вероятность в каждой точке.**

**CDF показывает, как накапливается вероятность при движении по оси x.**

Для дискретных распределений:

 - PMF показывает вероятность конкретных значений k .
 - CDF показывает накопленную вероятность до k .
Значения PMF и CDF различаются, но оба имеют смысл в дискретном контексте.

Для непрерывных распределений:

 - PDF показывает плотность вероятности.
 - CDF показывает накопленную вероятность.
 
Вероятность в точке для непрерывных распределений всегда равна 0, поэтому её не показывают отдельно.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Создаем фигуру с подграфиками
fig, axes = plt.subplots(6, 1, figsize=(12, 20))
axes = axes.flatten()

# Параметры для построения графиков
x_range = np.linspace(-3, 10, 1000)

# 1. Нормальное распределение
normal_dist = stats.norm(0, 1)
x_normal = np.linspace(-4, 4, 1000)
y_normal_pdf = normal_dist.pdf(x_normal)
y_normal_cdf = normal_dist.cdf(x_normal)
# PPF для некоторых вероятностей
prob_points = [0.025, 0.5, 0.975]
ppf_points = [normal_dist.ppf(p) for p in prob_points]
ppf_values = [normal_dist.pdf(x) for x in ppf_points]

axes[0].plot(x_normal, y_normal_pdf, 'b-', linewidth=2, label='PDF')
axes[0].plot(x_normal, y_normal_cdf, 'r--', linewidth=2, label='CDF')
axes[0].scatter(ppf_points, ppf_values, color='green', s=100, zorder=5, 
                label='PPF точки', marker='^')
axes[0].scatter(ppf_points, prob_points, color='orange', s=100, zorder=5,
                label='PPF значения (вероятности)', marker='s')
axes[0].axvline(0, color='purple', linestyle='--', linewidth=2, 
                label=f'PDF(0) = {normal_dist.pdf(0):.4f}')
axes[0].set_title('Нормальное распределение N(0,1)\nPPF: 2.5% = {ppf_points[0]:.2f}, 50% = {ppf_points[1]:.2f}, 97.5% = {ppf_points[2]:.2f}')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Плотность / Вероятность')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 2. Экспоненциальное распределение
exp_dist = stats.expon(scale=2)
x_exp = np.linspace(0, 10, 1000)
y_exp_pdf = exp_dist.pdf(x_exp)
y_exp_cdf = exp_dist.cdf(x_exp)
# PPF для некоторых вероятностей
prob_points_exp = [0.1, 0.5, 0.9]
ppf_points_exp = [exp_dist.ppf(p) for p in prob_points_exp]
ppf_values_exp = [exp_dist.pdf(x) for x in ppf_points_exp]

axes[1].plot(x_exp, y_exp_pdf, 'b-', linewidth=2, label='PDF')
axes[1].plot(x_exp, y_exp_cdf, 'r--', linewidth=2, label='CDF')
axes[1].scatter(ppf_points_exp, ppf_values_exp, color='green', s=100, zorder=5,
                label='PPF точки', marker='^')
axes[1].scatter(ppf_points_exp, prob_points_exp, color='orange', s=100, zorder=5,
                label='PPF значения (вероятности)', marker='s')
axes[1].axvline(1, color='purple', linestyle='--', linewidth=2, 
                label=f'PDF(1) = {exp_dist.pdf(1):.4f}')
axes[1].set_title('Экспоненциальное распределение (scale=2)\nPPF: 10% = {ppf_points_exp[0]:.2f}, 50% = {ppf_points_exp[1]:.2f}, 90% = {ppf_points_exp[2]:.2f}')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Плотность / Вероятность')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# 3. Равномерное распределение
uniform_dist = stats.uniform(0, 10)  # от 0 до 10
x_uniform = np.linspace(-1, 11, 1000)
y_uniform_pdf = uniform_dist.pdf(x_uniform)
y_uniform_cdf = uniform_dist.cdf(x_uniform)
# PPF для некоторых вероятностей
prob_points_unif = [0.1, 0.5, 0.9]
ppf_points_unif = [uniform_dist.ppf(p) for p in prob_points_unif]
ppf_values_unif = [uniform_dist.pdf(x) for x in ppf_points_unif]

axes[2].plot(x_uniform, y_uniform_pdf, 'b', linewidth=2, label='PDF')
axes[2].plot(x_uniform, y_uniform_cdf, 'r--', linewidth=2, label='CDF')
axes[2].scatter(ppf_points_unif, ppf_values_unif, color='green', s=100, zorder=5,
                label='PPF точки', marker='^')
axes[2].scatter(ppf_points_unif, prob_points_unif, color='orange', s=100, zorder=5,
                label='PPF значения (вероятности)', marker='s')
axes[2].axvline(5, color='purple', linestyle='--', linewidth=2, 
                label=f'PDF(5) = {uniform_dist.pdf(5):.4f}')
axes[2].set_title('Равномерное распределение U(0,10)\nPPF: 10% = {ppf_points_unif[0]:.2f}, 50% = {ppf_points_unif[1]:.2f}, 90% = {ppf_points_unif[2]:.2f}')
axes[2].set_xlabel('x')
axes[2].set_ylabel('Плотность / Вероятность')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

# 4. Гамма-распределение
gamma_dist = stats.gamma(2, scale=1)
x_gamma = np.linspace(0, 10, 1000)
y_gamma_pdf = gamma_dist.pdf(x_gamma)
y_gamma_cdf = gamma_dist.cdf(x_gamma)
# PPF для некоторых вероятностей
prob_points_gamma = [0.1, 0.5, 0.9]
ppf_points_gamma = [gamma_dist.ppf(p) for p in prob_points_gamma]
ppf_values_gamma = [gamma_dist.pdf(x) for x in ppf_points_gamma]

axes[3].plot(x_gamma, y_gamma_pdf, 'b', linewidth=2, label='PDF')
axes[3].plot(x_gamma, y_gamma_cdf, 'r--', linewidth=2, label='CDF')
axes[3].scatter(ppf_points_gamma, ppf_values_gamma, color='green', s=100, zorder=5,
                label='PPF точки', marker='^')
axes[3].scatter(ppf_points_gamma, prob_points_gamma, color='orange', s=100, zorder=5,
                label='PPF значения (вероятности)', marker='s')
axes[3].axvline(2, color='purple', linestyle='--', linewidth=2, 
                label=f'PDF(2) = {gamma_dist.pdf(2):.4f}')
axes[3].set_title('Гамма-распределение (a=2, scale=1)\nPPF: 10% = {ppf_points_gamma[0]:.2f}, 50% = {ppf_points_gamma[1]:.2f}, 90% = {ppf_points_gamma[2]:.2f}')
axes[3].set_xlabel('x')
axes[3].set_ylabel('Плотность / Вероятность')
axes[3].legend()
axes[3].grid(True, alpha=0.3)

# 5. Бета-распределение
beta_dist = stats.beta(2, 5)
x_beta = np.linspace(0, 1, 1000)
y_beta_pdf = beta_dist.pdf(x_beta)
y_beta_cdf = beta_dist.cdf(x_beta)
# PPF для некоторых вероятностей
prob_points_beta = [0.1, 0.5, 0.9]
ppf_points_beta = [beta_dist.ppf(p) for p in prob_points_beta]
ppf_values_beta = [beta_dist.pdf(x) for x in ppf_points_beta]

axes[4].plot(x_beta, y_beta_pdf, 'b', linewidth=2, label='PDF')
axes[4].plot(x_beta, y_beta_cdf, 'r--', linewidth=2, label='CDF')
axes[4].scatter(ppf_points_beta, ppf_values_beta, color='green', s=100, zorder=5,
                label='PPF точки', marker='^')
axes[4].scatter(ppf_points_beta, prob_points_beta, color='orange', s=100, zorder=5,
                label='PPF значения (вероятности)', marker='s')
axes[4].axvline(0.5, color='purple', linestyle='--', linewidth=2, 
                label=f'PDF(0.5) = {beta_dist.pdf(0.5):.4f}')
axes[4].set_title('Бета-распределение (α=2, β=5)\nPPF: 10% = {ppf_points_beta[0]:.3f}, 50% = {ppf_points_beta[1]:.3f}, 90% = {ppf_points_beta[2]:.3f}')
axes[4].set_xlabel('x')
axes[4].set_ylabel('Плотность / Вероятность')
axes[4].legend()
axes[4].grid(True, alpha=0.3)

# 6. Сводная таблица значений всех функций
axes[5].axis('off')  # Скрываем оси для таблицы

# Создаем таблицу с данными
data = [
    ['Распределение', 'Точка x', 'PDF(x)', 'CDF(x)', 'PPF(0.5)'],
    ['Нормальное', '0', f'{normal_dist.pdf(0):.4f}', f'{normal_dist.cdf(0):.4f}', f'{normal_dist.ppf(0.5):.4f}'],
    ['Экспоненциальное', '1', f'{exp_dist.pdf(1):.4f}', f'{exp_dist.cdf(1):.4f}', f'{exp_dist.ppf(0.5):.4f}'],
    ['Равномерное', '5', f'{uniform_dist.pdf(5):.4f}', f'{uniform_dist.cdf(5):.4f}', f'{uniform_dist.ppf(0.5):.4f}'],
    ['Гамма', '2', f'{gamma_dist.pdf(2):.4f}', f'{gamma_dist.cdf(2):.4f}', f'{gamma_dist.ppf(0.5):.4f}'],
    ['Бета', '0.5', f'{beta_dist.pdf(0.5):.4f}', f'{beta_dist.cdf(0.5):.4f}', f'{beta_dist.ppf(0.5):.4f}']
]

# Добавляем таблицу
table = axes[5].table(cellText=data[1:], colLabels=data[0], 
                     cellLoc='center', loc='center',
                     colWidths=[0.2, 0.15, 0.2, 0.2, 0.25])
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)
axes[5].set_title('Сводная таблица PDF, CDF, PPF значений', fontsize=14, pad=20)

plt.tight_layout()
plt.show()

# Выводим числовые значения в консоль
print("ЧИСЛОВЫЕ ЗНАЧЕНИЯ ВСЕХ ФУНКЦИЙ")
print("=" * 80)
print("1. НОРМАЛЬНОЕ РАСПРЕДЕЛЕНИЕ N(0,1):")
print(f"   PDF(0) = {normal_dist.pdf(0):.6f}")
print(f"   CDF(0) = {normal_dist.cdf(0):.6f}")
print(f"   PPF(0.5) = {normal_dist.ppf(0.5):.6f}")
print(f"   PPF(0.025) = {normal_dist.ppf(0.025):.6f} (2.5%)")
print(f"   PPF(0.975) = {normal_dist.ppf(0.975):.6f} (97.5%)")

print("\n2. ЭКСПОНЕНЦИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ (scale=2):")
print(f"   PDF(1) = {exp_dist.pdf(1):.6f}")
print(f"   CDF(1) = {exp_dist.cdf(1):.6f}")
print(f"   PPF(0.5) = {exp_dist.ppf(0.5):.6f}")

print("\n3. РАВНОМЕРНОЕ РАСПРЕДЕЛЕНИЕ U(0,10):")
print(f"   PDF(5) = {uniform_dist.pdf(5):.6f}")
print(f"   CDF(5) = {uniform_dist.cdf(5):.6f}")
print(f"   PPF(0.5) = {uniform_dist.ppf(0.5):.6f}")

print("\n4. ГАММА-РАСПРЕДЕЛЕНИЕ (a=2, scale=1):")
print(f"   PDF(2) = {gamma_dist.pdf(2):.6f}")
print(f"   CDF(2) = {gamma_dist.cdf(2):.6f}")
print(f"   PPF(0.5) = {gamma_dist.ppf(0.5):.6f}")

print("\n5. БЕТА-РАСПРЕДЕЛЕНИЕ (α=2, β=5):")
print(f"   PDF(0.5) = {beta_dist.pdf(0.5):.6f}")
print(f"   CDF(0.5) = {beta_dist.cdf(0.5):.6f}")
print(f"   PPF(0.5) = {beta_dist.ppf(0.5):.6f}")

# Демонстрация взаимосвязи PDF/CDF/PPF
print("ДЕМОНСТРАЦИЯ ВЗАИМОСВЯЗИ ФУНКЦИЙ")
print("Проверка: CDF(PPF(0.5)) = 0.5")
print(f"Нормальное: CDF(PPF(0.5)) = {normal_dist.cdf(normal_dist.ppf(0.5)):.10f}")
print(f"Экспоненциальное: CDF(PPF(0.5)) = {exp_dist.cdf(exp_dist.ppf(0.5)):.10f}")
print(f"Равномерное: CDF(PPF(0.5)) = {uniform_dist.cdf(uniform_dist.ppf(0.5)):.10f}")

print("\nПроверка: PPF(CDF(x)) = x")
test_x = 1.0
print(f"Нормальное: PPF(CDF({test_x})) = {normal_dist.ppf(normal_dist.cdf(test_x)):.10f}")
print(f"Экспоненциальное: PPF(CDF({test_x})) = {exp_dist.ppf(exp_dist.cdf(test_x)):.10f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Создаем фигуру с подграфиками
fig, axes = plt.subplots(6, 1, figsize=(15, 30))
axes = axes.flatten()

# Увеличиваем размер шрифтов глобально
plt.rcParams.update({
    'font.size': 14,
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 12,
    'figure.titlesize': 18
})

# 1. Биномиальное распределение
n, p = 10, 0.5
binomial_dist = stats.binom(n, p)
x_binom = np.arange(0, n+1)
y_binom_pmf = binomial_dist.pmf(x_binom)
y_binom_cdf = binomial_dist.cdf(x_binom)
# PPF для некоторых вероятностей
prob_points_binom = [0.05, 0.25, 0.5, 0.75, 0.95]
ppf_points_binom = [binomial_dist.ppf(p_val) for p_val in prob_points_binom]
ppf_pmf_values = [binomial_dist.pmf(int(x)) for x in ppf_points_binom]
ppf_cdf_values = [prob_points_binom[i] for i in range(len(prob_points_binom))]

axes[0].bar(x_binom, y_binom_pmf, alpha=0.7, color='blue', edgecolor='black', label='PMF', width=0.6)
axes[0].plot(x_binom, y_binom_cdf, 'r-', linewidth=3, marker='o', markersize=8, label='CDF')
# Добавляем PPF точки
axes[0].scatter(ppf_points_binom, ppf_pmf_values, color='green', s=150, zorder=5,
                label='PPF точки (PMF)', marker='^', edgecolors='black', linewidth=1)
axes[0].scatter(ppf_points_binom, prob_points_binom, color='orange', s=150, zorder=5,
                label='PPF вероятности', marker='s', edgecolors='black', linewidth=1)
axes[0].scatter(5, binomial_dist.pmf(5), color='red', s=200, zorder=5, edgecolors='black',
                label=f'PMF(5) = {binomial_dist.pmf(5):.4f}')
axes[0].scatter(5, binomial_dist.cdf(5), color='darkred', s=200, zorder=5, marker='^', edgecolors='black',
                label=f'CDF(5) = {binomial_dist.cdf(5):.4f}')
axes[0].set_title('Биномиальное распределение B(10, 0.5)\n'
                  f'PPF: 5%={ppf_points_binom[0]}, 25%={ppf_points_binom[1]}, '
                  f'50%={ppf_points_binom[2]}, 75%={ppf_points_binom[3]}, 95%={ppf_points_binom[4]}', 
                  fontsize=16, pad=20)
axes[0].set_xlabel('k (количество успехов)', fontsize=14)
axes[0].set_ylabel('Вероятность', fontsize=14)
axes[0].legend(fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(0, 1.05)

# 2. Пуассоновское распределение
lam = 3
poisson_dist = stats.poisson(lam)
x_poisson = np.arange(0, 15)
y_poisson_pmf = poisson_dist.pmf(x_poisson)
y_poisson_cdf = poisson_dist.cdf(x_poisson)
# PPF для некоторых вероятностей
prob_points_poisson = [0.1, 0.25, 0.5, 0.75, 0.9]
ppf_points_poisson = [poisson_dist.ppf(p_val) for p_val in prob_points_poisson]
ppf_pmf_values_poisson = [poisson_dist.pmf(int(x)) for x in ppf_points_poisson]
ppf_cdf_values_poisson = [prob_points_poisson[i] for i in range(len(prob_points_poisson))]

axes[1].bar(x_poisson, y_poisson_pmf, alpha=0.7, color='green', edgecolor='black', label='PMF', width=0.6)
axes[1].plot(x_poisson, y_poisson_cdf, 'r-', linewidth=3, marker='o', markersize=8, label='CDF')
# Добавляем PPF точки
axes[1].scatter(ppf_points_poisson, ppf_pmf_values_poisson, color='green', s=150, zorder=5,
                label='PPF точки (PMF)', marker='^', edgecolors='black', linewidth=1)
axes[1].scatter(ppf_points_poisson, prob_points_poisson, color='orange', s=150, zorder=5,
                label='PPF вероятности', marker='s', edgecolors='black', linewidth=1)
axes[1].scatter(2, poisson_dist.pmf(2), color='red', s=200, zorder=5, edgecolors='black',
                label=f'PMF(2) = {poisson_dist.pmf(2):.4f}')
axes[1].scatter(2, poisson_dist.cdf(2), color='darkred', s=200, zorder=5, marker='^', edgecolors='black',
                label=f'CDF(2) = {poisson_dist.cdf(2):.4f}')
axes[1].set_title('Пуассоновское распределение Pois(3)\n'
                  f'PPF: 10%={ppf_points_poisson[0]}, 25%={ppf_points_poisson[1]}, '
                  f'50%={ppf_points_poisson[2]}, 75%={ppf_points_poisson[3]}, 90%={ppf_points_poisson[4]}', 
                  fontsize=16, pad=20)
axes[1].set_xlabel('k (количество событий)', fontsize=14)
axes[1].set_ylabel('Вероятность', fontsize=14)
axes[1].legend(fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(0, 1.05)

# 3. Геометрическое распределение
p_geom = 0.3
geom_dist = stats.geom(p_geom)
x_geom = np.arange(1, 21)
y_geom_pmf = geom_dist.pmf(x_geom)
y_geom_cdf = geom_dist.cdf(x_geom)
# PPF для некоторых вероятностей
prob_points_geom = [0.1, 0.25, 0.5, 0.75, 0.9]
ppf_points_geom = [geom_dist.ppf(p_val) for p_val in prob_points_geom]
ppf_pmf_values_geom = [geom_dist.pmf(int(x)) for x in ppf_points_geom]
ppf_cdf_values_geom = [prob_points_geom[i] for i in range(len(prob_points_geom))]

axes[2].bar(x_geom, y_geom_pmf, alpha=0.7, color='orange', edgecolor='black', label='PMF', width=0.6)
axes[2].plot(x_geom, y_geom_cdf, 'r-', linewidth=3, marker='o', markersize=8, label='CDF')
# Добавляем PPF точки
axes[2].scatter(ppf_points_geom, ppf_pmf_values_geom, color='green', s=150, zorder=5,
                label='PPF точки (PMF)', marker='^', edgecolors='black', linewidth=1)
axes[2].scatter(ppf_points_geom, prob_points_geom, color='orange', s=150, zorder=5,
                label='PPF вероятности', marker='s', edgecolors='black', linewidth=1)
axes[2].scatter(3, geom_dist.pmf(3), color='red', s=200, zorder=5, edgecolors='black',
                label=f'PMF(3) = {geom_dist.pmf(3):.4f}')
axes[2].scatter(3, geom_dist.cdf(3), color='darkred', s=200, zorder=5, marker='^', edgecolors='black',
                label=f'CDF(3) = {geom_dist.cdf(3):.4f}')
axes[2].set_title('Геометрическое распределение Geom(0.3)\n'
                  f'PPF: 10%={ppf_points_geom[0]}, 25%={ppf_points_geom[1]}, '
                  f'50%={ppf_points_geom[2]}, 75%={ppf_points_geom[3]}, 90%={ppf_points_geom[4]}', 
                  fontsize=16, pad=20)
axes[2].set_xlabel('k (номер первого успеха)', fontsize=14)
axes[2].set_ylabel('Вероятность', fontsize=14)
axes[2].legend(fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].set_ylim(0, 1.05)

# 4. Гипергеометрическое распределение
M, n_hyper, N = 50, 20, 10  # всего, успехов, выборка
hypergeom_dist = stats.hypergeom(M, n_hyper, N)
x_hyper = np.arange(max(0, N-(M-n_hyper)), min(n_hyper, N)+1)
y_hyper_pmf = hypergeom_dist.pmf(x_hyper)
y_hyper_cdf = hypergeom_dist.cdf(x_hyper)
# PPF для некоторых вероятностей
prob_points_hyper = [0.1, 0.25, 0.5, 0.75, 0.9]
ppf_points_hyper = [hypergeom_dist.ppf(p_val) for p_val in prob_points_hyper]
ppf_pmf_values_hyper = [hypergeom_dist.pmf(int(x)) for x in ppf_points_hyper]
ppf_cdf_values_hyper = [prob_points_hyper[i] for i in range(len(prob_points_hyper))]

axes[3].bar(x_hyper, y_hyper_pmf, alpha=0.7, color='purple', edgecolor='black', label='PMF', width=0.6)
axes[3].plot(x_hyper, y_hyper_cdf, 'r-', linewidth=3, marker='o', markersize=8, label='CDF')
# Добавляем PPF точки
axes[3].scatter(ppf_points_hyper, ppf_pmf_values_hyper, color='green', s=150, zorder=5,
                label='PPF точки (PMF)', marker='^', edgecolors='black', linewidth=1)
axes[3].scatter(ppf_points_hyper, prob_points_hyper, color='orange', s=150, zorder=5,
                label='PPF вероятности', marker='s', edgecolors='black', linewidth=1)
axes[3].scatter(4, hypergeom_dist.pmf(4), color='red', s=200, zorder=5, edgecolors='black',
                label=f'PMF(4) = {hypergeom_dist.pmf(4):.4f}')
axes[3].scatter(4, hypergeom_dist.cdf(4), color='darkred', s=200, zorder=5, marker='^', edgecolors='black',
                label=f'CDF(4) = {hypergeom_dist.cdf(4):.4f}')
axes[3].set_title('Гипергеометрическое распределение Hyper(50,20,10)\n'
                  f'PPF: 10%={ppf_points_hyper[0]}, 25%={ppf_points_hyper[1]}, '
                  f'50%={ppf_points_hyper[2]}, 75%={ppf_points_hyper[3]}, 90%={ppf_points_hyper[4]}', 
                  fontsize=16, pad=20)
axes[3].set_xlabel('k (количество успехов в выборке)', fontsize=14)
axes[3].set_ylabel('Вероятность', fontsize=14)
axes[3].legend(fontsize=12)
axes[3].grid(True, alpha=0.3)
axes[3].set_ylim(0, 1.05)

# 5. Отрицательное биномиальное распределение
n_nb, p_nb = 5, 0.3
nbinom_dist = stats.nbinom(n_nb, p_nb)
x_nbinom = np.arange(0, 30)
y_nbinom_pmf = nbinom_dist.pmf(x_nbinom)
y_nbinom_cdf = nbinom_dist.cdf(x_nbinom)
# PPF для некоторых вероятностей
prob_points_nbinom = [0.1, 0.25, 0.5, 0.75, 0.9]
ppf_points_nbinom = [nbinom_dist.ppf(p_val) for p_val in prob_points_nbinom]
ppf_pmf_values_nbinom = [nbinom_dist.pmf(int(x)) for x in ppf_points_nbinom]
ppf_cdf_values_nbinom = [prob_points_nbinom[i] for i in range(len(prob_points_nbinom))]

axes[4].bar(x_nbinom, y_nbinom_pmf, alpha=0.7, color='brown', edgecolor='black', label='PMF', width=0.6)
axes[4].plot(x_nbinom, y_nbinom_cdf, 'r-', linewidth=3, marker='o', markersize=8, label='CDF')
# Добавляем PPF точки
axes[4].scatter(ppf_points_nbinom, ppf_pmf_values_nbinom, color='green', s=150, zorder=5,
                label='PPF точки (PMF)', marker='^', edgecolors='black', linewidth=1)
axes[4].scatter(ppf_points_nbinom, prob_points_nbinom, color='orange', s=150, zorder=5,
                label='PPF вероятности', marker='s', edgecolors='black', linewidth=1)
axes[4].scatter(10, nbinom_dist.pmf(10), color='red', s=200, zorder=5, edgecolors='black',
                label=f'PMF(10) = {nbinom_dist.pmf(10):.4f}')
axes[4].scatter(10, nbinom_dist.cdf(10), color='darkred', s=200, zorder=5, marker='^', edgecolors='black',
                label=f'CDF(10) = {nbinom_dist.cdf(10):.4f}')
axes[4].set_title('Отрицательное биномиальное распределение NB(5,0.3)\n'
                  f'PPF: 10%={ppf_points_nbinom[0]}, 25%={ppf_points_nbinom[1]}, '
                  f'50%={ppf_points_nbinom[2]}, 75%={ppf_points_nbinom[3]}, 90%={ppf_points_nbinom[4]}', 
                  fontsize=16, pad=20)
axes[4].set_xlabel('k (количество неудач до n успехов)', fontsize=14)
axes[4].set_ylabel('Вероятность', fontsize=14)
axes[4].legend(fontsize=12)
axes[4].grid(True, alpha=0.3)
axes[4].set_ylim(0, 1.05)

# 6. Сводная таблица значений PMF, CDF и PPF
axes[5].axis('off')  # Скрываем оси для таблицы

# Создаем таблицу с данными
data = [
    ['Распределение', 'Точка', 'PMF', 'CDF', 'PPF(0.5)'],
    ['Биномиальное', '5', f'{binomial_dist.pmf(5):.4f}', f'{binomial_dist.cdf(5):.4f}', f'{binomial_dist.ppf(0.5)}'],
    ['Пуассоновское', '2', f'{poisson_dist.pmf(2):.4f}', f'{poisson_dist.cdf(2):.4f}', f'{poisson_dist.ppf(0.5)}'],
    ['Геометрическое', '3', f'{geom_dist.pmf(3):.4f}', f'{geom_dist.cdf(3):.4f}', f'{geom_dist.ppf(0.5)}'],
    ['Гипергеометрическое', '4', f'{hypergeom_dist.pmf(4):.4f}', f'{hypergeom_dist.cdf(4):.4f}', f'{hypergeom_dist.ppf(0.5)}'],
    ['Отрицательное биномиальное', '10', f'{nbinom_dist.pmf(10):.4f}', f'{nbinom_dist.cdf(10):.4f}', f'{nbinom_dist.ppf(0.5)}']
]

# Добавляем таблицу
table = axes[5].table(cellText=data[1:], colLabels=data[0], 
                     cellLoc='center', loc='center',
                     colWidths=[0.2, 0.15, 0.2, 0.2, 0.25])
table.auto_set_font_size(False)
table.set_fontsize(14)
table.scale(1.5, 3)
axes[5].set_title('Сводная таблица PMF, CDF и PPF значений', fontsize=18, pad=30)

plt.tight_layout(pad=3.0)
plt.show()

# Выводим числовые значения в консоль
print("ЧИСЛОВЫЕ ЗНАЧЕНИЯ ВСЕХ ФУНКЦИЙ")
print("=" * 100)
print("1. БИНОМИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ B(10, 0.5):")
print(f"   PMF(5) = {binomial_dist.pmf(5):.6f}")
print(f"   CDF(5) = {binomial_dist.cdf(5):.6f}")
print(f"   PPF(0.5) = {binomial_dist.ppf(0.5)} (медиана)")
print(f"   PPF(0.05) = {binomial_dist.ppf(0.05)} (5-й процентиль)")
print(f"   PPF(0.95) = {binomial_dist.ppf(0.95)} (95-й процентиль)")

print("\n2. ПУАССОНОВСКОЕ РАСПРЕДЕЛЕНИЕ Pois(3):")
print(f"   PMF(2) = {poisson_dist.pmf(2):.6f}")
print(f"   CDF(2) = {poisson_dist.cdf(2):.6f}")
print(f"   PPF(0.5) = {poisson_dist.ppf(0.5)} (медиана)")

print("\n3. ГЕОМЕТРИЧЕСКОЕ РАСПРЕДЕЛЕНИЕ Geom(0.3):")
print(f"   PMF(3) = {geom_dist.pmf(3):.6f}")
print(f"   CDF(3) = {geom_dist.cdf(3):.6f}")
print(f"   PPF(0.5) = {geom_dist.ppf(0.5)} (медиана)")

print("\n4. ГИПЕРГЕОМЕТРИЧЕСКОЕ РАСПРЕДЕЛЕНИЕ Hyper(50,20,10):")
print(f"   PMF(4) = {hypergeom_dist.pmf(4):.6f}")
print(f"   CDF(4) = {hypergeom_dist.cdf(4):.6f}")
print(f"   PPF(0.5) = {hypergeom_dist.ppf(0.5)} (медиана)")

print("\n5. ОТРИЦАТЕЛЬНОЕ БИНОМИАЛЬНОЕ NB(5,0.3):")
print(f"   PMF(10) = {nbinom_dist.pmf(10):.6f}")
print(f"   CDF(10) = {nbinom_dist.cdf(10):.6f}")
print(f"   PPF(0.5) = {nbinom_dist.ppf(0.5)} (медиана)")

# Проверка, что сумма всех вероятностей = 1

print("ПРОВЕРКА СУММЫ ВЕРОЯТНОСТЕЙ (PMF)")
print(f"Сумма PMF биномиального: {np.sum(binomial_dist.pmf(x_binom)):.6f}")
print(f"Сумма PMF пуассоновского: {np.sum(poisson_dist.pmf(x_poisson)):.6f}")
print(f"Сумма PMF геометрического: {np.sum(geom_dist.pmf(x_geom)):.6f}")
print(f"Сумма PMF гипергеометрического: {np.sum(hypergeom_dist.pmf(x_hyper)):.6f}")
print(f"Сумма PMF отрицательного биномиального: {np.sum(nbinom_dist.pmf(x_nbinom)):.6f}")

# Проверка свойств CDF
print("\nПРОВЕРКА СВОЙСТВ CDF")
print(f"CDF(максимальное значение) для биномиального: {binomial_dist.cdf(n):.6f}")
print(f"CDF(максимальное значение) для пуассоновского: {poisson_dist.cdf(14):.6f}")
print(f"CDF(максимальное значение) для геометрического: {geom_dist.cdf(20):.6f}")

# Демонстрация взаимосвязи PMF/CDF/PPF

print("ДЕМОНСТРАЦИЯ ВЗАИМОСВЯЗИ ФУНКЦИЙ")
print("Проверка: CDF(PPF(0.5)) ≥ 0.5 (для дискретных распределений)")
print(f"Биномиальное: CDF(PPF(0.5)) = CDF({binomial_dist.ppf(0.5)}) = {binomial_dist.cdf(binomial_dist.ppf(0.5)):.6f}")
print(f"Пуассоновское: CDF(PPF(0.5)) = CDF({poisson_dist.ppf(0.5)}) = {poisson_dist.cdf(poisson_dist.ppf(0.5)):.6f}")

print("\nПроверка: PPF(CDF(x)) = x (приблизительно)")
test_x = 3
print(f"Биномиальное: PPF(CDF({test_x})) = PPF({binomial_dist.cdf(test_x):.6f}) = {binomial_dist.ppf(binomial_dist.cdf(test_x))}")
```
