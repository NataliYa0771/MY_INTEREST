### Центральная предельная теорема в действии: сходимость t-статистик к нормальному распределению ###

Основной вывод:

Проведенные эксперименты наглядно демонстрируют универсальность центральной предельной теоремы. Независимо от типа исходного распределения (нормальное, равномерное, экспоненциальное, биномиальное и др.), распределение t-статистик стремится к теоретическому нормальному распределению при достаточно большом размере выборки.

Наблюдаемая закономерность:

Нормальное распределение - t-статистики почти идеально совпадают с нормальным распределением (что ожидаемо)
Ненормальные распределения (равномерное, экспоненциальное, гамма и др.) - t-статистики также сходятся к нормальному распределению.

Статистические параметры: Среднее t-статистик близко к 0, стандартное отклонение близко к 1 (параметры стандартного нормального распределения)

Теоретическое обоснование:этот результат подтверждает центральную предельную теорему, согласно которой

 - Суммы (и, соответственно, средние) независимых случайных величин стремятся к нормальному распределению

 - t-статистика, будучи функцией разности средних, наследует это свойство

 - При больших объемах выборок (n ≥ 30) сходимость становится выраженной

Практическое значение:

**Данный эффект позволяет использовать t-тесты и другие статистические методы, основанные на нормальном распределении, даже когда исходные данные имеют произвольное распределение, при условии достаточного объема выборки.**

Эта функция генерирует выборки из заданного распределения, вычисляет среднее значение каждой выборки и рассчитывает t-статистику для сравнения двух выборок (тестовая и контрольная).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from tqdm import tqdm

def t_stats(distribution_func, params, sample_size=1000, M=1000):
    """
    Генерирует t-статистики для M пар тестовых и контрольных выборок.
    
    Параметры:
        distribution_func: Функция для генерации случайных чисел (например, np.random.normal).
        params: Параметры распределения (например, (mean, std) для нормального распределения).
        sample_size: Размер каждой выборки.
        M: Количество симуляций (пар тестовых и контрольных выборок).
    
    Возвращает:
        Список t-статистик для всех симуляций.
    """
    tstats = [] 

    for _ in tqdm(range(M)):
        # Генерация тестовой и контрольной выборок
        test_sample = distribution_func(*params, sample_size)
        control_sample = distribution_func(*params, sample_size)

        # Вычисление средних значений
        test_mean = np.mean(test_sample)
        control_mean = np.mean(control_sample)

        # Вычисление дисперсий с поправкой на степень свободы (ddof=1)
        test_var = np.var(test_sample, ddof=1)
        control_var = np.var(control_sample, ddof=1)

        # Вычисление стандартной ошибки разности средних
        sigma = np.sqrt(test_var / sample_size + control_var / sample_size)

        
        t = (test_mean - control_mean) / sigma
        tstats.append(t)

    return tstats

def plot_graph(distribution_func, params, sample_size=1000, M=1000):
    """
    Строит график сравнения теоретического и эмпирического распределений t-статистик.
    
    Параметры:
        distribution_func: Функция для генерации случайных чисел.
        params: Параметры распределения.
        sample_size: Размер каждой выборки.
        M: Количество симуляций.
    """
    # Получаем t-статистики из функции t_stats
    x_n = t_stats(distribution_func, params, sample_size, M)
    
    # Создаем ось x для теоретического нормального распределения
    x = np.linspace(-6, 6, 1000)

    # Рисуем теоретическое нормальное распределение
    plt.figure(figsize=(12, 8))
    plt.plot(x, stats.norm.pdf(x), 'red', lw=2, label="Theoretical Normal")

    # В зависимости от типа распределения рисуем KDE для t-статистик
    if distribution_func == np.random.normal:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Normal distribution')
        dist_name = "Normal"
    elif distribution_func == np.random.uniform:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Uniform distribution')
        dist_name = "Uniform"
    elif distribution_func == np.random.exponential:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Exponential distribution')
        dist_name = "Exponential"
    elif distribution_func == np.random.binomial:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Binomial distribution')
        dist_name = "Binomial"
    elif distribution_func == np.random.poisson:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Poisson distribution')
        dist_name = "Poisson"
    elif distribution_func == np.random.gamma:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Gamma distribution')
        dist_name = "Gamma"
    elif distribution_func == np.random.beta:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Beta distribution')
        dist_name = "Beta"
    elif distribution_func == np.random.chisquare:
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Chi-square distribution')
        dist_name = "Chi-square"
    else:
        # Для других распределений просто рисуем KDE
        sns.kdeplot(x_n, fill=True, alpha=0.6, label='Empirical distribution')
        dist_name = "Unknown"
    
    # Настройки графика
    plt.xlim(-6, 6)
    plt.xlabel('t-statistic')
    plt.ylabel('Density')
    plt.title(f'T-distribution comparison for {sample_size} random variables\n'
              f'{dist_name} vs Normal distribution\n'
              f'Mean: {np.mean(x_n):.4f}, Std: {np.std(x_n):.4f}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    print(f"\n=== Статистика для {dist_name} распределения ===")
    print(f"Среднее t-статистик: {np.mean(x_n):.4f}")
    print(f"Стандартное отклонение t-статистик: {np.std(x_n):.4f}")
    print(f"Количество симуляций: {len(x_n)}")
    print(f"Теоретическое значение для нормального распределения: 0.0 (среднее), 1.0 (стд)")

# Примеры использования для всех видов распределений:
if __name__ == "__main__":
    
    # 1. Нормальное распределение
    print("1. НОРМАЛЬНОЕ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.normal, (0, 1), sample_size=1000, M=1000)
    
    # 2. Равномерное распределение
    print("\n2. РАВНОМЕРНОЕ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.uniform, (0, 1), sample_size=1000, M=1000)
    
    # 3. Экспоненциальное распределение
    print("\n3. ЭКСПОНЕНЦИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.exponential, (1,), sample_size=1000, M=1000)
    
    # 4. Биномиальное распределение
    print("\n4. БИНОМИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.binomial, (100, 0.5), sample_size=1000, M=1000)
    
    # 5. Пуассоновское распределение
    print("\n5. ПУАССОНОВСКОЕ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.poisson, (3,), sample_size=1000, M=1000)
    
    # 6. Гамма-распределение
    print("\n6. ГАММА-РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.gamma, (2, 1), sample_size=1000, M=1000)
    
    # 7. Бета-распределение
    print("\n7. БЕТА-РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.beta, (2, 5), sample_size=1000, M=1000)
    
    # 8. Хи-квадрат распределение
    print("\n8. ХИ-КВАДРАТ РАСПРЕДЕЛЕНИЕ")
    plot_graph(np.random.chisquare, (5,), sample_size=1000, M=1000)
```
